```python
import pandas as pd
import numpy as np
```

Can not drop `NaN` or eslse not enough data available!

```python
data = pd.read_csv('owid_data.csv')
```

```python
data['country'].nunique()
```

```python
country_data = data[data['iso_code'] != np.nan]
```

Get needed columns

```python
country_data
```

```python
data = country_data[['country','year','gdp', 
                        'coal_electricity',
                        'gas_electricity', 
                        'nuclear_electricity', 
                        'oil_electricity',
                        'solar_electricity', 
                        'wind_electricity']].dropna()

top_50_gdp_countries = data.groupby('country').sum().sort_values('gdp', ascending=False).iloc[1:31].index

electricity_data = data[data['country'].isin(top_50_gdp_countries)]

electricity_data = electricity_data.reset_index(drop=True)

electricity_data
```

```python
min_year = electricity_data.groupby('country')['year'].min()

growth_gdp = electricity_data.groupby('country').apply(lambda df: df['gdp'] / df[df['year'] == df['year'].min()]['gdp'].iloc[0]).reset_index()['gdp']
# growth_elec = electricity_data.groupby('country').apply(lambda df: df['electricity_generation'] / df[df['year'] == df['year'].min()]['electricity_generation'].iloc[0]).reset_index()['electricity_generation']

electricity_data['growth_gdp'] = growth_gdp
electricity_data = electricity_data.drop(columns='gdp')
```

```python
electricity_data.to_csv('electricity_data1.csv')
```
